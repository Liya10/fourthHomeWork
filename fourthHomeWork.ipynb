```python
 import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'    
df = pd.read_csv(url, sep = '\t')
print("What is the number of observations in the dataset?", df.index.size)#4621
```

```python
print("What is the number of columns in the dataset? ",len(df.columns))
```

```python
print("Which was the most-ordered item and how many times it was ordered? ",np.argmax(np.bincount(df['order_id'])))
```

```python
print("How many different items are sold?",len(set(df['item_name'])))
```

```python
print("How many items were orderd in total?",np.sum(df['quantity']))
```

```python
print(" Turn the item price into a float")
df['item_price']=df['item_price'].replace('[\$,]','',regex=True).astype(float)
```

```python
print(" What is the price of each item?")
df.head(20)
```

```python
print("How much was the revenue for the period in the dataset?", np.sum(df['item_price']))
```

```python
print("How many orders were made in the period?",np.max(df['order_id']))
```

```python
print("What is the average revenue amount per order?", df['item_price'].mean())
```

```python
print("Merge the table dfa that contains order date and order id with df and plot daily revenue")
dfa = pd.DataFrame({"date": np.random.choice(pd.date_range(start="20170101", end="20170201"), size=1835),
                   "order_id": np.arange(1835)})
df2=pd.merge(df, dfa, on='order_id')
df2.head(20)
```

```python
print("Create a histogram of the top 5 items bought")
d=df.groupby('item_name').sum().sort_values(by='quantity').tail(5).index

for i in range(5):
    tmp=df[df['item_name']==d[i]]
    plt.bar(tmp['order_id'],tmp['item_price']) 
    plt.legend((tmp['item_name']), loc=(1.04, 0.4))
    plt.xlabel(u'order_id')
    plt.ylabel(u'item_price')
    plt.show()
```

```python
print("Sort the dataset by the name of the item.")
d=df2.sort_values(by='item_name')
d
```

```python
print('How many times were a Veggie Salad Bowl ordered?',len(df[df['item_name']=='Veggie Salad Bowl']))
```

```python
print('How many times people orderd more than one Canned Soda?')
print(len(df[(df['item_name']=='Canned Soda')&(df['quantity']>1)]))
```

```python
print('Create a scatterplot with the number of items orderered per order price')

df2=df.groupby('item_price').sum()

plt.scatter( df2.index,df2['quantity'], color="#000066")
plt.xlabel(u'Цена')
plt.ylabel(u'Количество заказанных товаров по данной цене')
plt.show()
```
